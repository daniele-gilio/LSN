## Exercise 3.1

In this exercise we were asked to evaluate the price of European call and put options. We first evaluated those by sampling directly the price at time $T=1$. Then we tried to evaluate said price by following it's evolution over dicretized times. As we expected the values obtained are within $2\sigma$'s from the analytical result (represented by the coninuous horizontal lines).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, put, error = np.loadtxt("direct.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.errorbar(block,put,yerr=error,marker='s')
plt.axhline(y=5.4595325819072364)
plt.title("Put's direct extimation as a function of number of blocks")
plt.xlabel('Block Number')
plt.ylabel('Put Value')
plt.grid(1)
plt.rcParams['figure.figsize'] = [16,9]
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, call, error = np.loadtxt("direct.dat", usecols=(0,3,4), delimiter=";", unpack='true')
plt.errorbar(block,call,yerr=error,marker="s")
plt.axhline(y=14.975790778311286)
plt.title("Call's direct extimation as a function of number of blocks")
plt.xlabel('Block Number')
plt.ylabel('Call Value')
plt.grid(1)
plt.show()
```
