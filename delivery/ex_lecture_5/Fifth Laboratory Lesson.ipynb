# Exercise 5

In this exercise we were asked to implement the Metropolis algorithm to sample Hydrogen atoms' 1s and 1p eigenstates. The plots below show the points' distribution in 3D space and the average radius of the eigenstates (in $a_0$ terms), computed with a Uniform Distribution and a Gaussian Distribution. The results are as expected.  

## Uniform Distribution

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =  np.loadtxt("test_run_r0.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [16,9]
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.title("S State Distribution")
plt.show()
```

```python
X, Y, Z =  np.loadtxt("test_run_r1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.title("P State Distribution")
plt.show()
```

```python
block, r0, error = np.loadtxt("r0.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.errorbar(block,r0,yerr=error)
plt.axhline(y=1.5)
plt.title("Mean Radius (S State)")
plt.xlabel('Block Number')
plt.ylabel('r')
plt.grid(1)
plt.show()
```

```python
block, r1, error = np.loadtxt("r1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.errorbar(block,r1,yerr=error)
plt.axhline(y=5)
plt.title("Mean Radius (P State)")
plt.xlabel('Block Number')
plt.ylabel('r')
plt.grid(1)
plt.show()
```
