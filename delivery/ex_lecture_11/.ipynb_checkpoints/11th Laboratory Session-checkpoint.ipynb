# Excercise 11.1
In this excercise we had to fit a linear function via a neural network. To do this, we had to optimize the hyperparameters of the simplest neural network possible, one with only one neuron. This kind of network is well capable of optimizing a linear function because it matches its number of parameters. All we did to make the network converge was reducing the batch size to 4.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 100)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=4, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

# Excercise 11.2
In this section we implemented a different neural network to fit the following polynomial function:
$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.
<br>
To make this network converge we increased the number of both train and validation data, used an Elu activation function and changed the optimizer to Adam. Our network consists in an sequence of 5 identical layers, followed by an output layer to match the output dimension. We also added an Early Stopping callback for convenience.

```python
#polinomial parameters
a = 3.
b=-2.
c=-3.
d=4.
def poly(x):
    return d+c*x+b*x**2.+a*x**3.


# generate training inputs
#np.random.seed(0)
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target = poly(x_valid) # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(poly(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(poly(x_valid), sigma)

# plot training and target dataset
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model

model = tf.keras.Sequential()

n_layer = 4

model.add(Dense(4, input_shape=(1,),activation='elu'))

for i in range (n_layer):
    model.add(Dense(4,activation='elu'))

model.add(Dense(1))
# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()

ES = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
#RLR = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1)
```

```python
# fit the model using training dataset
# over 30 epochs of 8 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=8, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid), callbacks=[ES])

```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(1)
plt.ylim(0,2)
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

# Excercise 11.3

In this section we had to expand the previous work to fit $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$. We basically used the same network, adjusting it to accept a bidimensional tensor as an input. We also implemented a Reduce Learning Rate On Plateau and changed the activation function to a Selu to increase performance.

```python
def s(x):
    return np.sin(x[:,0]**2.+x[:,1]**2.)

# generate training inputs
np.random.seed(0)
n_data = 10000
n_validation = int(n_data*0.1)
x_train = np.zeros((n_data,2))
x_valid = np.zeros((n_validation,2))
x_train[:,0] = np.random.uniform(-1.5, 1.5, n_data)
x_train[:,1] = np.random.uniform(-1.5,1.5,n_data)
x_valid[:,0] = np.random.uniform(-1.5, 1.5, n_validation)
x_valid[:,1] = np.random.uniform(-1.5, 1.5, n_validation)

#y_target=np.zeros((2,n_data))
y_target = s(x_valid) # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
#y_train=np.zeros((2,n_data))
y_train = np.random.normal(s(x_train), sigma) # actual measures from which we want to guess regression parameters
#y_valid=np.zeros((2,n_data))
y_valid = np.random.normal(s(x_valid), sigma)


from mpl_toolkits.mplot3d import Axes3D

X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_target
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, marker="o", color="grey", label="Validation Set")

X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_valid
ax.plot(X, Y, Z, marker='+',color="g", label="Theoretical", linestyle="None")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
#ax.view_init(10, 30)
plt.legend()
plt.show()

```

```python
# compose the NN model

model = tf.keras.Sequential()

n_layer = 4

#x_train=x_train.transpose()
#x_valid=x_valid.transpose()
#y_train.reshape(1,10000)
#y_valid.reshape(1,1000)

model.add(Dense(4,input_shape=(2,),activation='selu'))

for i in range (n_layer):
    model.add(Dense(4,activation='selu'))

model.add(Dense(1))
# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='adam', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()

ES = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
RLR = ReduceLROnPlateau(monitor='val_loss', factor=0.7,patience=2, verbose=1)
print(x_train.shape, y_train.shape)
print(x_valid.shape, y_valid.shape)
```

```python
# fit the model using training dataset
# over 30 epochs of 8 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=8, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid), callbacks=[ES,RLR])
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(1)
plt.show()
```

```python
x_predicted=np.zeros((1000,2))
x_predicted[:,0] = np.random.uniform(-1.5, 1.5, 1000)
x_predicted[:,1] = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = model.predict(x_predicted)

X = x_predicted[:,0]
Y = x_predicted[:,1]
Z = y_predicted
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, marker="o", color="b", label="Predicted")
X = x_valid[:,0]
Y = x_valid[:,1]
Z = y_target
ax.plot(X, Y, Z, marker='+',color="r", linestyle="None", label="Thoretic",markevery=6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.legend()
plt.show()
```
