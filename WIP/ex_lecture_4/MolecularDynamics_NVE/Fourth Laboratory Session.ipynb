### Argon

### Solid Phase

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, value, error = np.loadtxt("ave_pot_argon_1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.rcParams['figure.figsize'] = [26,21]
plt.subplot(3,2,1)
plt.errorbar(block,value,yerr=error,marker="+")
plt.title("Average Potential Energy per Particle")
plt.xlabel('Block Number')
plt.ylabel('U/N [J]')
plt.grid(1)
block, value, error = np.loadtxt("ave_kin_argon_1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.subplot(3,2,2)
plt.errorbar(block,value,yerr=error,marker="+")
plt.title("Average Kinetic Energy per Particle")
plt.xlabel('Block Number')
plt.ylabel('K/N [J]')
plt.grid(1)
block, value, error = np.loadtxt("ave_etot_argon_1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.subplot(3,2,3)
plt.errorbar(block,value,yerr=error,marker="+")
plt.title("Average Total Energy per Particle")
plt.xlabel('Block Number')
plt.ylabel('(U+V)/N [J]')
plt.grid(1)
block, value, error = np.loadtxt("ave_temp_argon_1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.subplot(3,2,4)
plt.errorbar(block,value,yerr=error,marker="+")
plt.title("Average Temperature")
plt.xlabel('Block Number')
plt.ylabel('T [K]')
plt.grid(1)
block, value, error = np.loadtxt("ave_p_argon_1.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.subplot(3,2,5)
plt.errorbar(block,value,yerr=error,marker="+")
plt.title("Average Pressure")
plt.xlabel('Block Number')
plt.ylabel('P [atm]')
plt.grid(1)
plt.show()
```
