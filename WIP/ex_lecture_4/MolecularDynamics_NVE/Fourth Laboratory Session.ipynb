```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

block, I, error = np.loadtxt("ave_pot.dat", usecols=(0,1,2), delimiter=";", unpack='true')
plt.errorbar(block,I,yerr=error,marker="+")
plt.title("Potential Energy")
plt.rcParams['figure.figsize'] = [16,9]
plt.xlabel('Block Number')
plt.ylabel('U (Lennard-Jones Units)')
plt.grid(1)
plt.show()
```
