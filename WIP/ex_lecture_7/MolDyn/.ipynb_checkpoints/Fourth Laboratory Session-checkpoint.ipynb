# Excercise 4

In excercise 4 we were asked to modify a provided Molecular Dynamics. The first request was to implement a restart mechanism to let the sistem get to equilibrium before taking proper measurements. The second request was to add data blocking to measure average properties and their statistical errors. Then we had to add some code to measure pressure, exploiting the virial theorem. Lastly we did measurements of solid,liquid and gas phases of Argon and Krypton. In reality we just had to make 3 simulations, since we used the corresponding states theorem.

## Excercise 4.1 and 4.3

We can clearly see from the graphs below that the restart mechanism, obtained by rescaling velocities, works flawlessly with the first generic phase test. Real elements measures required 2 to 4 restarts before reaching the desired temperature. All measurements are expressed in Lennard-Jones units.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16,9]
temp = np.loadtxt("output_temp_generic.dat", usecols=(0), unpack='true')
plt.plot(temp, label="Temperature")
u = np.loadtxt("output_epot_generic.dat", usecols=(0), unpack='true')
plt.plot(u, label="Potential Energy")
k = np.loadtxt("output_ekin_generic.dat", usecols=(0), unpack='true')
plt.plot(k, label="Kinetic Energy")
t = np.loadtxt("output_etot_generic.dat", usecols=(0), unpack='true')
plt.plot(t, label="Total Energy")
p = np.loadtxt("output_press_generic.dat", usecols=(0), unpack='true')
plt.plot(p, label="Pressure")
plt.title("Excercise 4.1")
plt.xlabel('#Measure')
plt.ylabel('Value')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

## Excercise 4.2 and 4.3

In the graphs below we added data blocking, the trends of the errors are as expected. Note that we're still using the generic phase, whose values are stored in "input.dat". All the measurements are expressed in Lennard-Jones Units.

```python
block,temp,error = np.loadtxt("ave_temp_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, temp, yerr=error, label="Temperature")
plt.title("Excercise 4.2")
plt.xlabel('Block Number')
plt.ylabel('Temperature')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

```python
block,k,error = np.loadtxt("ave_ekin_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, k, yerr=error, label="Kinetic Energy")
plt.title("Excercise 4.2")
plt.xlabel('Block Number')
plt.ylabel('Kinetic Energy')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

```python
block,p,error = np.loadtxt("ave_press_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, p, yerr=error, label="Pressure")
plt.title("Excercise 4.2")
plt.xlabel('Block Number')
plt.ylabel('Pressure')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

```python
block,u,error = np.loadtxt("ave_epot_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, u, yerr=error, label="Potential Energy")
plt.title("Excercise 4.2")
plt.xlabel('Block Number')
plt.ylabel('Potential Energy')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```

```python
block,t,error = np.loadtxt("ave_etot_generic.dat", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(block, t, yerr=error, label="Total Energy")
plt.title("Excercise 4.2")
plt.xlabel('Block Number')
plt.ylabel('Total Energy')
plt.grid(1)
plt.legend(loc="best")
plt.show()
```
