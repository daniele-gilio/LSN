## Excercise 8.1 and 8.2
In this section we show graphs of the Ground State Energy and Wave Function Probability $|\psi|^2$ obtained by means of our code, which exploits the Variational Monte Carlo. The first step of the process is to optimize the chosen wave function parameters (we chose a sum of gaussian functions symmetric with respect to the origin and with the same $\sigma$). We than use the Metropolis Algorithm to sample the wave function and compute the energy. We can see in the graphs below both the Ground State Energy and the sampled Wave Function Probability compared with the test function, evaluated with optimized parameters, and with a wave function probability obtained exploiting matrix algebra.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("potential.dat", delimiter=";")
x=k[:,0]
y=k[:,1]
err=k[:,2]

plt.errorbar(x,y,yerr=err, marker=".")
plt.title("Data Blocking Ground State Energy")
plt.xlabel("Block Number")
plt.ylabel("Energy")
plt.grid()
plt.show()
```

```python
def psi_t(x,mu,sigma):
    return (np.exp(-(x-mu)**2./(2.*sigma**2.))+np.exp(-(x+mu)**2./(2.*sigma**2.)))/np.sqrt(2.*np.sqrt(np.pi)*sigma*(1.+np.exp(-(mu/sigma)**2.)))

mu_opt=0.79
sigma_opt=0.62

k=np.loadtxt("hist.dat", delimiter=";")
y=k[:,0]
err=k[:,1]
x=np.linspace(-5.,5.,len(y))
plt.errorbar(x,y, yerr=err, label="Numerical Wave Function Histogram")

for i in range(len(y)):
    y[i]=psi_t(x[i],mu_opt,sigma_opt)*psi_t(x[i],mu_opt,sigma_opt)

plt.plot(x,y,label="Test Function with Optimized Parameters")

#Lecture Code
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
#plt.figure(figsize=(8,5))
#scale = 0.3
plt.plot(x,(psi[0])**2, label="Matrix Function")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.title("Squared Modulus of $\psi$ as a Function of x")

plt.grid()
plt.legend()
plt.show()
```

## Excercise 8.3
Here we can find a comparison between our code wave function, one obtained with PIGS (T=0) and one obtained with PIMC (T>0). The latter takes into consideration a mixed quantum state since the absolute temperature is more than zero, so excited states are accessible via thermal agitation.

```python
k=np.loadtxt("QMC_1D/probability.dat")
x=k[:,0]
y=k[:,1]
err=k[:,2]

plt.errorbar(x,y,yerr=err, label="QMC Function (PIGS)")

k=np.loadtxt("hist.dat", delimiter=";")
y=k[:,0]
err=k[:,1]
x=np.linspace(-5.,5.,len(y))
plt.errorbar(x,y, yerr=err, label="Numerical Wave Function Histogram")

k=np.loadtxt("QMC_1D/probability_pimc.dat")
x=k[:,0]
y=k[:,1]
err=k[:,2]

plt.errorbar(x,y,yerr=err, label="QMC Function (PIMC)")

plt.xlim(-5.,5.)
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid()
plt.legend()
plt.show()
```
