## Circumference Distribution
In this section we show cost and path resuslts for both the serial and the parallel iterations of the program. As we expected they both get to the same path and the parallel cost is "lower" than the serial one.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("serial/best_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Cost Serial")

print("Best Serial Cost ", y[-1])

k=np.loadtxt("parallel/best_cost_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Cost Parallel")

print("Best Parallel Cost ", y[-1])

plt.axhline(y=2*np.pi,xmin=0,xmax=100)

plt.axhline(y=2*np.pi,xmin=0,xmax=100, label="Expected Value")
plt.title("Cost Function for Circumference")
plt.xlabel("Temperature")
plt.ylabel("Cost")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [14,14]

k=np.loadtxt("serial/best_path_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path Serial")

k=np.loadtxt("serial/best_path_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path Parallel")

x=np.linspace(-1.,1., 10000)
y=np.sqrt(1.-x*x)

plt.plot(x,y, label="Upper Circumference")

y=-np.sqrt(1.-x*x)

plt.plot(x,y, label="Lower Circumference")
plt.title("Best Path for Circumference")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

## Square Area Distribution
As in the circumference distribution we get that both iterations of the program coverge to a similar solution costwise. We cannot evaluate the path from the graph since we don't know the solution. The parallel cost overcomes the serial one but it gets "lower" a few points later.

```python
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("serial/best_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Cost Serial")

print("Best Serial Cost ", y[-1])

k=np.loadtxt("parallel/best_cost_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path Parallel")

print("Best Parallel Cost ", y[-1])

plt.title("Cost Function for Square Area")
plt.xlabel("Temperature")
plt.ylabel("Cost")
plt.grid()
plt.legend()
plt.show()
```

```python
k=np.loadtxt("serial/best_path_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path Serial")

k=np.loadtxt("parallel/best_path_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path Parallel")

plt.title("Best Path for Square Area")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```
