```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("best_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".")
plt.axhline(y=2*np.pi,xmin=0,xmax=100)
plt.title("Cost Function for Circumference")
plt.xlabel("Generation")
plt.ylabel("Cost")
plt.grid()
plt.show()
```

```python
k=np.loadtxt("best_path_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".")

x=np.linspace(-1.,1., 10000)
y=np.sqrt(1.-x*x)

plt.plot(x,y)

y=-np.sqrt(1.-x*x)

plt.plot(x,y)
plt.title("Best Path for Circumference")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
