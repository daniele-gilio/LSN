```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("best.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", linestyle="none")
plt.title("Data Blocking Ground State Energy")
plt.xlabel("Block Number")
plt.ylabel("Energy")
plt.grid()
plt.show()
```
