# Excercise 9
In this excercise we were asked to create a code which functioned by means of a genetic algorithm. The purpose of the program is to solve the Travelling Salesman problem. Genetic algorithms are very susceptible to parameters selection, so my optimization (done via trial and error) is not universal and may not be the optimal in terms of both computational cost and results. We show below two instances of the problem. The first one has the cities distributed on a circumference; this functions as a test, since we know that the miminum path lenght should be $2\pi$ (we took a circumference of radius $1$). The second one has a more interesting problem, the cities are arranged randomly on a square area of side lenght equal to $1$. To write the code we needed to implement a selection algorithm and some mutations. I also implemented a "mercy mechanism" which saves individuals picking them randomly (but giving more probability to the best ones).

## Circumference Distribution
There's nothing particularly interesting about circumference's graphs. The results are clearly in line with what's expected.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("best_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Cost")
plt.axhline(y=2*np.pi,xmin=0,xmax=100)

k=np.loadtxt("mean_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Mean Cost")
plt.axhline(y=2*np.pi,xmin=0,xmax=100, label="Expected Value")
plt.title("Cost Function for Circumference")
plt.xlabel("Generation")
plt.ylabel("Cost")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [14,14]

k=np.loadtxt("best_path_circ.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Path")

x=np.linspace(-1.,1., 10000)
y=np.sqrt(1.-x*x)

plt.plot(x,y, label="Upper Circumference")

y=-np.sqrt(1.-x*x)

plt.plot(x,y, label="Lower Circumference")
plt.title("Best Path for Circumference")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

## Square Area Distribution
We cannot evaluate the "goodness" of this instance of the program since we don't know the answer "a priori". We can see an interesting pattern comparing the mean and the best cost. When the best cost is constant the mean varies quite a lot, meaning that mutations, breeding and the mercy mechanism are working correctly. Increasing the generation number results in a constant best, which means that we found a family of equivalent solutions.

```python
plt.rcParams['figure.figsize'] = [16,9]

k=np.loadtxt("best_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Best Cost")

k=np.loadtxt("mean_sq.dat", delimiter=" ")
x=k[:,0]
y=k[:,1]

plt.plot(x,y, marker=".", label="Mean Cost")

plt.title("Cost Function for Square Area")
plt.xlabel("Generation")
plt.ylabel("Cost")
plt.grid()
plt.legend()
plt.show()
```
